```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_data():
    data_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_459954.csv'
    country_metadata_path = 'Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_459954.csv'
    indicator_metadata_path = 'Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_459954.csv'

    data = pd.read_csv(data_path, skiprows=4)
    country_metadata = pd.read_csv(country_metadata_path)
    indicator_metadata = pd.read_csv(indicator_metadata_path)

    return data, country_metadata, indicator_metadata

data, country_metadata, indicator_metadata = load_data()
```

```python
def merge_data(year):
    data, country_metadata, _ = load_data()

    # Select relevant columns and merge datasets
    population_data = data[['Country Code', 'Country Name', str(year)]]
    merged_data = pd.merge(population_data, country_metadata[['Country Code', 'Region', 'IncomeGroup']], on='Country Code')

    # Handle missing data
    merged_data[str(year)].fillna(0, inplace=True)

    return merged_data

year = 2020
merged_data = merge_data(year)
```

```python
# Display the merged data
print("Merged Data Head:\n", merged_data.head())
```

```python
def plot_population_distribution(year, region=None, income_group=None, chart_type='bar', save=False, export_data=False):
    merged_data = merge_data(year)
```

```python
# Group by region and sum the population
region_population = merged_data.groupby('Region')[str(year)].sum().reset_index()
```

```python
# Plot the chart
plt.figure(figsize=(12, 6))

# Set a default value for chart_type
chart_type = 'bar'

if chart_type == 'bar':
    bars = plt.bar(region_population['Region'], region_population[str(year)], color='skyblue')
    plt.xlabel('Region')
    plt.ylabel(f'Population in {year}')
    plt.title(f'Population Distribution by Region in {year}')
    plt.xticks(rotation=45)
```

```python
def plot_bar_chart(year):
    data, country_metadata, _ = load_data()
    population_data = data[['Country Name', str(year)]].dropna()

    plt.figure(figsize=(10, 6))
    plt.bar(population_data['Country Name'][:10], population_data[str(year)][:10], color='skyblue')
    plt.xlabel('Country')
    plt.ylabel('Population')
    plt.title(f'Population in {year}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plot_bar_chart(2020)
```

```python
def plot_pie_chart(year):
    data, country_metadata, _ = load_data()
    population_data = data[['Country Name', str(year)]].dropna()

    plt.figure(figsize=(8, 8))
    plt.pie(population_data[str(year)][:10], labels=population_data['Country Name'][:10], autopct='%1.1f%%', startangle=140)
    plt.title(f'Population Distribution in {year}')
    plt.show()

plot_pie_chart(2020)
```

```python
import seaborn as sns

def plot_line_chart(countries, start_year, end_year):
    data, _, _ = load_data()
    years = [str(year) for year in range(start_year, end_year + 1)]
    filtered_data = data[data['Country Name'].isin(countries)][['Country Name'] + years]

    melted_data = filtered_data.melt(id_vars=['Country Name'], var_name='Year', value_name='Population')
    melted_data['Year'] = melted_data['Year'].astype(int)

    plt.figure(figsize=(12, 6))
    sns.lineplot(x='Year', y='Population', hue='Country Name', data=melted_data, marker='o')
    plt.title('Population Over Time')
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.legend(title='Country')
    plt.show()

plot_line_chart(['United States', 'China', 'India'], 2000, 2020)
```

```python
def plot_scatter_chart(year):
    data, _, _ = load_data()
    population_data = data[['Country Name', str(year)]].dropna()

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=population_data['Country Name'][:10], y=population_data[str(year)][:10])
    plt.xlabel('Country')
    plt.ylabel('Population')
    plt.title(f'Population Scatter Plot in {year}')
    plt.xticks(rotation=45)
    plt.show()

plot_scatter_chart(2020)
```

```python
def plot_histogram(year):
    data, _, _ = load_data()
    population_data = data[str(year)].dropna()

    plt.figure(figsize=(10, 6))
    plt.hist(population_data, bins=30, color='skyblue', edgecolor='black')
    plt.xlabel('Population')
    plt.ylabel('Frequency')
    plt.title(f'Population Histogram in {year}')
    plt.show()

plot_histogram(2020)
```

```python
import seaborn as sns

def plot_box_plot(year):
    data, _, _ = load_data()
    population_data = data[['Country Name', str(year)]].dropna()

    plt.figure(figsize=(12, 6))
    sns.boxplot(x=population_data[str(year)])
    plt.xlabel('Population')
    plt.title(f'Population Distribution in {year}')
    plt.show()

plot_box_plot(2020)
```

```python
def plot_heatmap(years):
    data, _, _ = load_data()
    years = [str(year) for year in years]
    population_data = data[['Country Name'] + years].dropna()

    population_data.set_index('Country Name', inplace=True)

    plt.figure(figsize=(12, 8))
    sns.heatmap(population_data, cmap='viridis')
    plt.title('Population Heatmap')
    plt.show()

plot_heatmap([2000, 2005, 2010, 2015, 2020])
```

```python
def plot_pair_plot(years):
    data, _, _ = load_data()
    years = [str(year) for year in years]
    population_data = data[['Country Name'] + years].dropna()

    population_data_melted = population_data.melt(id_vars=['Country Name'], var_name='Year', value_name='Population')

    sns.pairplot(population_data_melted, hue='Country Name', kind='scatter')
    plt.title('Population Pair Plot')
    plt.show()

plot_pair_plot([2000, 2005, 2010, 2015, 2020])
```

```python
def plot_violin_plot(year):
    data, _, _ = load_data()
    population_data = data[['Country Name', str(year)]].dropna()

    plt.figure(figsize=(12, 6))
    sns.violinplot(x=population_data[str(year)])
    plt.xlabel('Population')
    plt.title(f'Population Violin Plot in {year}')
    plt.show()

plot_violin_plot(2020)
```

```python
def plot_strip_plot(year):
    data, _, _ = load_data()
    population_data = data[['Country Name', str(year)]].dropna()

    plt.figure(figsize=(12, 6))
    sns.stripplot(x=population_data[str(year)])
    plt.xlabel('Population')
    plt.title(f'Population Strip Plot in {year}')
    plt.show()

plot_strip_plot(2020)
```
